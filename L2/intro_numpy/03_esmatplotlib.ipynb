```python
# Importa le librerie necessarie
import matplotlib.pyplot as plt
import numpy as np

# Imposta Matplotlib per visualizzare i grafici inline nel notebook
%matplotlib inline

# Esempio 1: Grafico semplice
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure()  # Crea una nuova figura
plt.plot(x, y, label='sin(x)')  # Plot di sin(x)
plt.title('Grafico di sin(x)')  # Titolo del grafico
plt.xlabel('x')  # Etichetta asse x
plt.ylabel('sin(x)')  # Etichetta asse y
plt.legend()  # Mostra la legenda
plt.grid(True)  # Mostra la griglia
plt.show()  # Mostra il grafico
```

```python
#esempio 2
# Esempio 2: Scatter plot con colori e dimensioni diverse
rng = np.random.RandomState(0) #specifichiamo un "seme" di inizio, cos ogni volta che rieseguiamo il codice avremo gli stessi numeri casuali. Utile per riproducibilit risultati
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
plt.figure()
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.colorbar()  # Aggiungi barra dei colori
plt.title('Scatter plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Esempio 3: Istogramma
data = np.random.randn(1000)
plt.figure()
#plt.plot(np.arange(1000),data)
plt.hist(data, bins=30, edgecolor='black')
plt.title('Istogramma dei dati')
plt.xlabel('Valore')
plt.ylabel('Frequenza')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generazione dei dati per le distribuzioni
# Uniforme
uniform_data = np.random.uniform(low=0, high=1, size=1000)

# Normale (Gaussiana)
normal_data = np.random.normal(loc=0, scale=3, size=1000)

# Creazione dei grafici
plt.figure(figsize=(12, 6))

# Plot per la distribuzione Uniforme
plt.subplot(1, 2, 1)
plt.hist(uniform_data, bins=30, edgecolor='black')
plt.title('Distribuzione Uniforme')
plt.xlabel('Valori casuali')
plt.ylabel('Frequenza')

# Plot per la distribuzione Normale (Gaussiana)
plt.subplot(1, 2, 2)
plt.hist(normal_data, bins=30, edgecolor='black', color='orange')
plt.title('Distribuzione Normale (Gaussiana)')
plt.xlabel('Valori casuali')
plt.ylabel('Frequenza')

# Mostra il grafico
plt.tight_layout()
plt.show()
```

```python
# Dati per il grafico 3D
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Plot del grafico 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')
ax.set_title('Grafico 3D di sin(sqrt(x^2 + y^2))')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
