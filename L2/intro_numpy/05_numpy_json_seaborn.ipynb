```python
import json
import numpy as np
```

```python
city_position = None
with open("data/cities.json") as f:
    city_position = json.load(f)
```

```python
city_position
```

```python
list_of_coordinates=[]
city_position_length = len(city_position)
array_of_coordinates = np.array([])
array_of_names = np.array([])

for idx, c_el in enumerate(city_position):
    coords = np.array([c_el["lat"], c_el["lon"]])
    array_of_names = np.append(array_of_names, c_el["city"])
    list_of_coordinates = np.append(list_of_coordinates,coords)

print(array_of_names)
list_of_coordinates
```

```python
matrix_el = list_of_coordinates.reshape(
    int(len(list_of_coordinates)/2), 2
)
matrix_el
```

```python
from sklearn.metrics import pairwise_distances
dist_matrix = pairwise_distances(matrix_el)
```

```python
import seaborn as sns
ax = sns.heatmap(dist_matrix)
ax.set_xticklabels( array_of_names, rotation=90 )
ax.set_yticklabels( array_of_names, rotation=0 )
```
