```python
import numpy as np
import matplotlib.pyplot as plt
import PIL.Image as Image
```

```python
im = Image.open("./data/cat.jpg")
im = im.convert("L")

#im.save("prova.png","L")
X = np.reshape(np.matrix(im.getdata()), (im.size[1],im.size[0]))
print("X.shape =", X.shape)
plt.figure(1); plt.imshow(X,'gray') 
```

```python
[U,S,V]=np.linalg.svd(X,full_matrices=0)
S = np.diag(S)
i = 0
for k in [700, 200, 50, 30, 10]:
  i = i+1
  m = U.shape[0]; n = V.shape[1]
  plt.figure(i); plt.imshow(U[:,0:k]*S[0:k,0:k]*V[0:k,:],'gray')
  plt.title('fattore di compressione = 1 : ' + str(float(m*n)/float(m*k+k+k*n)))
  print('fattore di compressione = 1 : ', float(m*n)/float(m*k+k+k*n))
#endfor
```
