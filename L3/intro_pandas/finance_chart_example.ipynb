```python
import matplotlib.pyplot as plt
from mplfinance.original_flavor import candlestick_ohlc
import pandas as pd
import numpy as np
import matplotlib.dates as mdates

# Dati per il grafico a candele (esempio di dati finanziari)
data = pd.read_csv('../data/AAPL.csv')  # Sostituisci con il percorso del tuo file CSV
data['Date'] = pd.to_datetime(data['Date'])
data['Date'] = data['Date'].apply(mdates.date2num)

ohlc = data[['Date', 'Open', 'High', 'Low', 'Close']].values

# Plot del grafico a candele
fig, ax = plt.subplots(figsize=(10, 6))
candlestick_ohlc(ax, ohlc, width=0.6, colorup='g', colordown='r')
ax.xaxis_date()
ax.set_title('Grafico a Candele di Dati Finanziari')
ax.set_xlabel('Data')
ax.set_ylabel('Prezzo')
plt.xticks(rotation=45)
plt.show()
```
