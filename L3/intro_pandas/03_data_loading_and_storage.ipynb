Se ho un dataset molto ampio potrebbe essere utile per efficienza spezzarlo, si fa con il parametro chunksize di read_csv. Questo approccio evita di caricare l'intero file CSV in memoria e permette di gestire file di grandi dimensioni in modo efficiente. Puo essere utile anche per parallelizzare calcoli con multiprocessing un chunk per processo

```python
sars_data_chunked = pd.read_csv("../data/sars_2003_complete_dataset.csv", chunksize=100)
sars_data_chunked # restituisce un iteratore
```

```python
total_sum = 0
for chunk in sars_data_chunked:
    total_sum += chunk['deaths'].sum()
    print("somma cumulata chunk per chunk: ", total_sum)
print(f"Somma totale: {total_sum}")
```
